```python
import nibabel as nib
import matplotlib.pyplot as plt


def plot_nifti(nifti_file_path):
    img = nib.load(nifti_file_path)

    data = img.get_fdata()

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    axes[0].imshow(data[:, :, data.shape[2] // 2])
    axes[0].set_title("Axial Plane")

    axes[1].imshow(data[:, data.shape[1] // 2, :])
    axes[1].set_title("Coronal Plane")

    sagittal_slice = data[data.shape[0] // 2, :, :]
    axes[2].imshow(sagittal_slice, origin="lower")
    axes[2].set_title("Sagittal Plane")

    plt.show()
```

```python
# Scan Without Preprocessing

nifti_file_path = "./data/train/IBSR_01/IBSR_01.nii.gz"
plot_nifti(nifti_file_path)
```

```python
# Scan With Preprocessing

nifti_file_path = "./data/train/IBSR_01/IBSR_01_5labels_preprocessed.nii.gz"
plot_nifti(nifti_file_path)
```

```python
# Ground Truth Without Preprocessing

nifti_file_path = "./data/train/IBSR_01/IBSR_01_seg.nii.gz"
plot_nifti(nifti_file_path)
```

```python
# Ground Truth With Preprocessing

nifti_file_path = "./data/train/IBSR_01/IBSR_01_5labels_preprocessed_seg.nii.gz"
plot_nifti(nifti_file_path)
```
